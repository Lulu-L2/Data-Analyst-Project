Analisis pengidap penyakit jantung berdasarkan umur

```python
df.keys()
```

```python
fig,ax = plt.subplots(figsize=(12,6),dpi=150)
sns.countplot(df['age'],hue=df['target'],palette='winter')

ax.legend(loc='upper right');
```

# Target=1, Mengindikasikan bahwa pasien mengidap penyakit jantung

### Sebaran umur terhadap  pengidap penyakit jantung sangat banyak, mulai dari yang berumur 29 tahun

analisis pengidap penyakit jantung berdasarkan maximum heart rate (detakan jantung)

```python
fig,ax = plt.subplots(figsize=(12,6),dpi=150)

sns.barplot(x=df['age'],y=df['thalach'],hue=df['target'])
ax.set(ylabel='Maximum Heart Rate')

ax.legend(loc='upper right');
```

analisis pengidap penyakit jantung berdasarkan kolestrol level (detakan jantung)

```python
fig,ax = plt.subplots(figsize=(12,5),dpi=150)

sns.barplot(x=df['age'],y=df['chol'],hue=df['target'])

ax.legend(loc='upper left');
```

```python
fig,ax = plt.subplots(figsize=(12,6),dpi=150)

df.plot(ax=ax,kind='scatter',x='thalach',y='chol',c='target',cmap='Set2',figsize=(12,5))

ax.set(title='Correlation between max heart rate(thalach) and cholestrol level to heart disease');
```

# Kesimpulan
## Semakin tinggi kadar kolestrol dan semakin tinggi maximum detak jantung patient, maka akan semakin berpotensi mengidap penyakit jantung

# Analisis pengidap penyakit jantung pada pasien yang berumur diatas 50 tahun

```python
over_50=df[df['age']>50]
over_50
```

## Melihat sebaran max heart rate terhadap penyakit jantung pada pasien diatas 50 tahun

```python
fig,ax=plt.subplots(figsize=(12,5),dpi=150)

over_50.plot(ax=ax,kind='scatter',x='age',y='thalach',c='target',cmap='Set1');
```

# Melihat hubungan data yang saling mempengaruhi satu sama lain

```python
fig,ax=plt.subplots(figsize=(8,5),dpi=100)
sns.heatmap(ax=ax,data=over_50.corr());
```

# Visualisasi data yang mempengaruhi penyakit jantung pada pasien

```python
fig,(ax1,ax2)=plt.subplots(figsize=(10,10),dpi=100,nrows=2,ncols=1,sharex=True)

# Visualisasi hubungan antara umur dengan kolestrol level terhadap penyakit jantung,
# serta melihat rata-rata kolestrol level pasien pada data

# plot ax1
scatter=ax1.scatter(x=over_50['age'],y=over_50['chol'],c=over_50['target'],cmap='winter')
ax1.axhline(over_50['chol'].mean(),ls='--')

# customize ax1
ax1.set(xlim=(50,80),ylabel='cholestrol',title='Heart Disease and Cholestrol Levels')
ax1.legend(*scatter.legend_elements(),title='target')

# Visualisasi hubungan antara umur dengan max detak jantung terhadap penyakit jantung,
# serta melihat rata-rata detak jantung pasien pada data

# plot ax2
scatter=ax2.scatter(x=over_50['age'],y=over_50['thalach'],c=over_50['target'],cmap='winter')
ax2.axhline(over_50['thalach'].mean(),ls='--')

# customize ax2
ax2.set(ylabel='max heart rate',ylim=(60,200),xlabel='age',title='Heart Disease and Max Heart Rate')
ax2.legend(*scatter.legend_elements(),title='target')

# beri judul pada figure dan save figure ke local
fig.suptitle("Heart Disease Analysis",fontsize=16,fontweight='bold');
plt.savefig('heart-disease-analysis-plot-saved-with-code.png')
```
